```python
import numpy as np
import matplotlib.pyplot as plt
import imgprofiler as profiler
```

```python
#simulate an image with sin waves
line=np.sin(np.linspace(0.0,2.0*np.pi*10.0,100))
img=np.repeat(line[np.newaxis,:],100,axis=0)
img.shape
```

```python
plt.imshow(img)
```

```python
linecoords=np.array([[0.0,20.0,20.0],[0.0,20.0,60.0],[0.0,60.0,60.0]],dtype=float)
profile=profiler.getThickLineProfile(np.array([img]),linecoords[:2],1.0,zcoord1=0)
plt.plot(profile)
```

```python
polyprofile=profiler.getThickPolylineProfile(np.array([img]),linecoords,1.0,zcoord1=0)
```

```python
plt.plot(polyprofile)
```
